# Tratamiento de valores Faltantes


Vamos a llenar los valores nulos con un 'Sin Dato' porque en su mayoria son variables categóricas. Si fuera un dataframe con muchas variables numéricas, podriamos elegir reemplazarlas con la media, la mediana, o con un valor simple cómo cero/uno. Tambien podes considerar si es viable eliminarlos, los pros/contras de hacerlos.

##### Plataforma Amazon

```python
#A traves del metodo 'fillna' es que rellenamos los valores nulos
#Utilizamos el parámetro 'inplace = True' para que los cambios queden definitivos.
amazon_df.fillna('Sin Dato', inplace=True)
amazon_df.head()
```

```python
#Vemos que ahora no tenemos nulos, ya que los rellenamos
amazon_df.isna().sum()
```

##### Plataforma Disney Plus

```python
disney_df.fillna('Sin Dato', inplace = True)
disney_df.head()
```

```python
disney_df.isna().sum()
```

##### Plataforma Hulu

```python
#Rellenamos los valores faltantes con el 'Sin Dato' como venimos haciendo.
hulu_df.fillna('Sin Dato', inplace = True)
hulu_df.head()
```

```python
hulu_df.isna().sum()
```

##### Plataforma Netflix

```python
netflix_df.fillna('Sin Dato', inplace = True)
netflix_df.head()
```

```python
netflix_df.isna().sum()
```

#### CONCATENACION DEL DATAFRAME

```python
#Utilizamos el metodo concat para unir los dataframes, tienen que tener las mismas columnas para que se haga.
films_df = pd.concat([amazon_df,disney_df, hulu_df, netflix_df])
films_df
```

```python
#Vamos a separar mediante 'split' la parte cuantitativa de duracion de su unidad de tiempo(min/seasons)
ff = films_df['duracion'].str.split(" ", n = 1, expand = True)
films_df['duracion_film'] = ff[0]
films_df['unidad_duracion'] = ff[1]
films_df
```

```python
#Dropeamos la columna duracion que era la original, porque ya la separamos
films_df.drop(columns='duracion',inplace = True)
```

```python
#Renombramos la columna
films_df.rename(columns={'duracion_film' : 'duracion'}, inplace=True)
```

```python
#Notamos que hay 482 filas que tienen el valor 'Sin' en esta columna
films_df.duracion.value_counts()
```

```python
#Reemplazamos ese valor 'Sin' por 1, para que no influya en nuestros resultados
films_df.duracion.replace({"Sin": 1}, inplace= True)
films_df.duracion.value_counts()
```

```python
films_df
```

#### Guardamos nuestro dataframe final

```python
#Vamos a exportar nuestro dataframe terminado a csv, con el metodo 'to_csv'
films_df.to_csv('Datasets/films_df_final.csv')
```
